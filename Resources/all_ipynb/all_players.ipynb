```python
# file directory
file_dir = "C:/Users/mamoo/UT_DataScience/UTA-VIRT-DATA-PT-02-2021-U-B/FinalProject/Resources/"

# import the all_player.csv
all_players_df = pd.read_csv(f"{file_dir}all_players_2004to2020.csv")
pd.set_option("display.max_columns", None) # To show all columns explicitly
all_players_df
```

```python
# Check all the column headers.
all_players_df.columns
```

Most of the column headers are in abbreviated form which is hard to understand. We need to rename columns with a little elaborate titles.

```python
all_players_df = all_players_df.rename(columns={"POS": "Position", 
                               "GP": "GamesPlayed",
                              "GS": "GamesStarted",
                              "MINS": "MinutesPlayed",
                              "G": "GoalsScored",
                              "A": "Assists",
                              "SHTS": "Shots",
                              "SOG": "ShotsOnGoal",
                              "GWG": "GameWinningGoal",
                              "PKG/A": "PenaltyKicks",
                              "HmG": "HomeGoals",
                              "RdG": "RoadGoals",
                              "G/90min": "GoalsPer90Minutes",
                              "SC%": "ShotConversion",
                              "GWA": "GameWinningAssists",
                              "HmA": "HomeAssists",
                              "RdA": "RoadAssists",
                              "A/90min": "AssistsPer90Min",
                              "FC": "FoulsCommitted",
                              "FS": "FoulsSufferred",
                              "OFF": "Offsides",
                              "YC": "YellowCards",
                              "RC": "RedCards",
                              "SOG%": "ShotsOnGoalPercent",
                              "Year": "Year",
                              })
all_players_df.head(2)
```

```python
# Show any rows where 'Club' title is empty
all_players_df[(all_players_df['Club'] == '') | (all_players_df['Club'].isnull())]
```

So there are 4280 rows out of 4971 rows where the *Club* column has no value. In our analysis, the "Club" information is crucial. So we may drop the rows where there is no info for *Club*. 

```python
all_players_df[(all_players_df['Club'].isnull())]
```

```python
all_players_df = all_players_df.drop(all_players_df[all_players_df['Club'].isnull()].index)
all_players_df
```

```python
all_players_df['Club'].nunique()
```

```python
all_players_df['Club'].unique()
```

```python
all_players_df.isnull().sum()
```

It looks like the **ShotsOnGoalPercent** column has 151 null values (out of 691 rows).

The values under *ShotsOnGoalPercent* = *ShotsOnGoal* / *Shots*. Since this column is derived from mathematical manipulation of other two columns, if we drop this column the information should still be there in the dataframe.

```python
# Drop the column "ShotsOnGoalPercent"
all_players_df = all_players_df.drop("ShotsOnGoalPercent", axis='columns')
all_players_df.head(2)
```

```python
# The "PenaltyKicks" column has also some inconsistent data. We're focused on the team or club.
# As we still have 28 columns of data, we decide to to drop this column.
all_players_df = all_players_df.drop("PenaltyKicks", axis='columns')
all_players_df.head(2)
```

```python
all_players_df.describe()
```

```python
all_players_df.info()
```

The **all_players_df** is now **cleaned** with total of **26 columns, 691 rows** of non-null data with consistent data types.
